```python
# import modules
import urllib.request
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
%matplotlib inline
```

```python
# Load the JARVIS dataset
url = 'https://ndownloader.figshare.com/files/14146319' 
urllib.request.urlretrieve(url, 'jff2.json') 
f=open('jff2.json','r')
data=json.load(f)
f.close()

# Show the first item
data[0]
```

```python
# Create the space group array
sg_array = []
for d in data:
    if d['sg'] is not '':
        sg_array.append(d['sg'])
    else:
        sg_array.append('')

# Count the number of points for each space group
sg_hist = Counter()
for sg in sg_array:
     sg_hist[sg] += 1

# Sort the space groups from the largest number of points to the smalest
sg_sorted = [item[0] for item in sorted(sg_hist.items() , reverse=True, key=lambda x: x[1])]

# Plot the corresponding histogram
plt.rcParams['figure.figsize'] = [15, 12]
plt.title("Number of data point corresponding to each space group")
sns.set(style="darkgrid")
ax = sns.countplot(y=sg_array, order = sg_sorted)
```

```python
# Set Bravais lattices corresponding to the space groups explored above
Bravais_lattices = {'cubic' : {'Fm-3m','Im-3m','Pm-3m','Fd-3m','Pm-3n','Ia-3d','F-43m','P-43m','I-43m','Fm3'},
                    'hexagonal' : {'P6_3/mmc','P6/mmm','P6_3mc','P6_322'},
                    'tetragonal' : {'I4/mmm','P4/mmm','P4_2/mnm','I4/mcm','I4/m','I4_1/amd','I-4'},
                    'triclinic' : {'P-1','P1'},
                    'rhombohedral' : {'R-3m','P-3m1'},
                    'orthorhombic' : {'Cmce','Pnma','Cmmm','Cmcm','Pnnm','Immm','Pmmn','Fmmm','Cmme','Pbcm','P2_12_12_1'},
                    'monoclinic' : {'P2/m','C2/m','P2_1/c','P2_1/m','Pm','Pc','C2/c','P2/c','Cc','P2_1'},
                    'trigonal' : {'P3m1','R-3'}}

# define function retuning Bravais lattice corresponding to specified space group
def get_lattice(space_group):
    result = np.nan
    for bl in Bravais_lattices.items():
        if space_group in bl[1]:
            result = bl[0]
    return result

# Create the Bravais lattice array
bl_array = []
for d in data:
    bl_array.append(get_lattice(d['sg']))

# Count the number of points for each Bravais lattice
bl_hist = Counter()
for bl in bl_array:
     bl_hist[bl] += 1            
            
# Sort the Bravais lattices from the largest number of points to the smalest
bl_sorted = [item[0] for item in sorted(bl_hist.items() , reverse=True, key=lambda x: x[1])]         
            
# Plot the corresponding histogram
plt.rcParams['figure.figsize'] = [15, 5]
plt.title("Number of data point corresponding to each Bravais lattice")
sns.set(style="darkgrid")
ax = sns.countplot(x=bl_array, order=bl_sorted)
```

```python
# determine number of surfaces computed for materials with cubic lattices
sn_array = [len(d['surf_name']) for d in data if get_lattice(d['sg']) == 'cubic']

# Count the number of points for each number of surfaces
sn_hist = Counter()
for sn in sn_array:
     sn_hist[sn] += 1            
            
# Sort the Bravais lattices from the largest number of points to the smalest
sn_sorted = [item[0] for item in sorted(sn_hist.items() , reverse=True, key=lambda x: x[1])]         
            
# Plot the corresponding histogram
plt.rcParams['figure.figsize'] = [15, 5]
plt.title("Number of data point corresponding to each nuber of surfaces for materials with cubic lattices")
sns.set(style="darkgrid")
ax = sns.countplot(x=sn_array, order=sn_sorted)
```

```python
# determine number of vacancy energies computed for materials with cubic lattices and 13 surfaces
ve_array = [len(d['defect_val']) for d in data if get_lattice(d['sg']) == 'cubic' and len(d['surf_name']) == 13]

# Count the number of points for each number of surfaces
ve_hist = Counter()
for ve in ve_array:
     ve_hist[ve] += 1            
            
# Sort the Bravais lattices from the largest number of points to the smalest
ve_sorted = [item[0] for item in sorted(ve_hist.items() , reverse=True, key=lambda x: x[1])]         
            
# Plot the corresponding histogram
plt.rcParams['figure.figsize'] = [15, 5]
plt.title("Number of data point corresponding to number of vacancy energies computed for materials with cubic lattices and 13 surfaces")
sns.set(style="darkgrid")
ax = sns.countplot(x=ve_array, order=ve_sorted)
```

```python
# show all the materials with 1 positive vacancy energy
import pprint
pprint.pprint([{d['formula'] : d['defect_val']}  for d in data 
               if get_lattice(d['sg']) == 'cubic' 
               and len(d['surf_name']) == 13 
               and len(d['defect_val']) == 1 
               and d['defect_val'][0] > 0.0])
```

```python
# check the number of unique surfaces
surfaces = set()
for d in data: 
    if get_lattice(d['sg']) == 'cubic' and len(d['surf_name']) == 13 and len(d['defect_val']) == 1 and d['defect_val'][0] > 0.0:
        for s in d['surf_name']:
            surfaces.add(s) 
print(surfaces)    
```

```python
# create a pandas dataframe
df_cubic = pd.DataFrame(columns=['Potential','1_0_0 surface energy','1_1_0 surface energy','1_1_1 surface energy',
                                 '2_1_0 surface energy','2_1_1 surface energy','2_2_1 surface energy',
                                 '3_1_0 surface energy','3_1_1 surface energy','3_2_0 surface energy',
                                 '3_2_1 surface energy','3_2_2 surface energy','3_3_1 surface energy',
                                 '3_3_2 surface energy','c_1_1 elastic constant','c_1_2 elastic constant',
                                 'c_1_3  elastic constant', 'c_4_4  elastic constant','Shear modulus',
                                 'Bulk modulus','Cohesive energy','Vacancy formation energy'])

j = 0
for d in data: 
    if d['sg'] == 'Fm-3m' and len(d['surf_name']) == 13 and len(d['defect_val']) == 1 and d['defect_val'][0] > 0.0 and d['Bv'] > 0.0 and d['surf_val'][d['surf_name'].index('1_0_0')] > 0.0:
        df_cubic.loc[j] = [d['ff'], d['surf_val'][d['surf_name'].index('1_0_0')], d['surf_val'][d['surf_name'].index('1_1_0')], d['surf_val'][d['surf_name'].index('1_1_1')], 
                         d['surf_val'][d['surf_name'].index('2_1_0')], d['surf_val'][d['surf_name'].index('2_1_1')], d['surf_val'][d['surf_name'].index('2_2_1')],
                         d['surf_val'][d['surf_name'].index('3_1_0')], d['surf_val'][d['surf_name'].index('3_1_1')], d['surf_val'][d['surf_name'].index('3_2_0')],
                         d['surf_val'][d['surf_name'].index('3_2_1')], d['surf_val'][d['surf_name'].index('3_2_2')], d['surf_val'][d['surf_name'].index('3_3_1')],
                         d['surf_val'][d['surf_name'].index('3_3_2')], d['c11'], d['c12'], 
                         d['c13'], d['c44'], d['Gv'], d['Bv'], -d['fenp'], d['defect_val'][0]]
        j += 1

df_cubic.to_csv('data_fcc.csv')
print(j)
```
